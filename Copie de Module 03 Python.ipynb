Continue dans While Loop

```
# Skip odd numbers
x = 6
while x:
	x -= 1
	if x % 2 != 0:
		continue
	print(x)
```

Else dans While Loop

```
x = 6
while x:
    print(x)
    x -= 1
else:
    print('Done!')
```

**Python autorise une clause else facultative à la fin d'une boucle while. La clause else sera exécutée lorsque la boucle se terminera normalement (la condition devient fausse).**

La clause else sera toujours exécutée si la condition est fausse au début.

```
x = 0
while x:
    print(x)
    x -= 1
else:
    print('Done!')
```

Si la boucle se termine avec break, la clause else ne sera pas exécutée.

```
x = 6
while x:
    print(x)
    x -= 1
    if x == 3:
        break
else:
    print('Done!')
```

**Une compréhension List est un moyen compact de créer une structure de données Python à partir d'itérateurs. Avec les compréhensions, vous pouvez combiner des boucles et des tests conditionnels avec une syntaxe simplifier.**

[<the_expression> for <the_element> in <the_iterable>]

Exemples Basiques :

```
multiples = []
 
for n in range(1,16):
    multiples.append(n*5)
     
print(multiples)
```

```
multiples = [n*5 for n in range(1,16)]
 
print(multiples)
```

```
cubes = [n**3 for n in range(1,16)]
 
print(cubes)
```

**Conditions in List Comprehensions**

	
[<the_expression> for <the_element> in <the_iterable> if <the_condition>]

```
paires = []
 
for n in range(1,21):
    if n%2 == 0:
        paires.append(n)
         
print(paires)
```

```
paires = [n for n in range(1,21) if n%2 == 0]
 
print(paires)
```

[<the_expression> if <the_condition> else <other_expression> for <the_element> in <the_iterable>]

```
squares_cubes = []
 
for n in range(1,16):
    if n%2 == 0:
        squares_cubes.append(n**2)
    else:
        squares_cubes.append(n**3)
         
print(squares_cubes)
```

```
squares_cubes = [n**2 if n%2 == 0 else n**3 for n in range(1,16)]
 
print(squares_cubes)
```

**Access Nested List Items by Index**

```
L = ['a', 'b', ['cc', 'dd', ['eee', 'fff']], 'g', 'h']

print(L[2])
# Prints ['cc', 'dd', ['eee', 'fff']]

print(L[2][2])
# Prints ['eee', 'fff']

print(L[2][2][0])
# Prints eee
```

**Negative List Indexing In a Nested List**
![Texte alternatif…](https://www.learnbyexample.org/wp-content/uploads/python/Python-Nested-List-Negative-Indexing.png)

```
L = ['a', 'b', ['cc', 'dd', ['eee', 'fff']], 'g', 'h']

print(L[-3])
# Prints ['cc', 'dd', ['eee', 'fff']]

print(L[-3][-1])
# Prints ['eee', 'fff']

print(L[-3][-1][-2])
# Prints eee
```

**Change Nested List Item Value**

```
L = ['a', ['bb', 'cc'], 'd']
L[1][1] = 0
print(L)
# Prints ['a', ['bb', 0], 'd']
```

**Add items to a Nested list**

```
L = ['a', ['bb', 'cc'], 'd']
L[1].append('xx')
print(L)
# Prints ['a', ['bb', 'cc', 'xx'], 'd']
```

When you want to insert an item at a specific position in a nested list, use insert() method.

```
L = ['a', ['bb', 'cc'], 'd']
L[1].insert(0,'xx')
print(L)
# Prints ['a', ['xx', 'bb', 'cc'], 'd']
```

You can merge one list into another by using extend() method.

```
L = ['a', ['bb', 'cc'], 'd']
L[1].extend([1,2,3])
print(L)
# Prints ['a', ['bb', 'cc', 1, 2, 3], 'd']
```

**Remove items from a Nested List**

```
L = ['a', ['bb', 'cc', 'dd'], 'e']
x = L[1].pop(1)
print(L)
# Prints ['a', ['bb', 'dd'], 'e']

# removed item
print(x)
# Prints cc
```

If you don’t need the removed value, use the del statement.

```
L = ['a', ['bb', 'cc', 'dd'], 'e']
del L[1][1]
print(L)
# Prints ['a', ['bb', 'dd'], 'e']
```

```
L = ['a', ['bb', 'cc', 'dd'], 'e']
L[1].remove('cc')
print(L)
# Prints ['a', ['bb', 'dd'], 'e']
```
